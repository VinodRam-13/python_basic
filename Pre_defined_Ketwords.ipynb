 **Pre Define Keywords**



*   Keywords are pre defined words that holds a specific meaning and purpose
*   THey define structure and syntax of the language, enabling control flow, data type and various functionalities.

1.   **Control Flow**

     *   Keyword like if, else, elif, while and more are to control the flow of execution in program.






```python
x = 10
if x > 5:
  print("x is greater then 5")
else:
  print("x is not greater than 5 or equal to 5")
```


2.   **Logical operation**

     *   Keyword like and, or and not are used to perform logical operations.





```python
a = True
b = False
if a and b:
  print("a and b are true")
else:
  print("a and b are not true")
```



3.   **Exeception Handling**

      *   Keywords such as try, except, finally and raise are used for execption handing.





```python
try:
  result = 10/0
  print(result)
except ZeroDivisionError:
  print("Error: Division by zero")
finally:
  print("This will always be executed")
```
