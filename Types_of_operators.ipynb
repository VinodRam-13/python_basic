**Types of Operators**


*   Operators are special symbols or keywords that are used to carry out specific actions on numbers or variables in Python expressions.
*   There are several Operators in python language they are

    *   Arthimetic Operator
    *   Comparison Operator

    *   Logical Operator
    *   Assignment Operator

    *  Membership Operator
    *  Identity Operator

    *   Bitwise Operator

**Arthimetic Operator**


*   The Arthimetic Operator used to perform mathematical operations in programming
*   They are Addition, Subraction, Multiplication, Division, Floor, Modules, Exponention.













```python
a = 18
b = 7

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
floor_division = a // b
modulus = a % b
exponentiation = a ** b

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation)
```

**Comparison Operator**


*   Comparison Operator is used compare the two values and return a boolen value
*   They are ==, !=, <, >, >=, <=



```python
x = 18
y = 7

print(x > y)
print(x == y)
print(x != y)
print(x <= y)
```

**Logical Operator**


*   Logical Operator are used to combine conditional statement
*   They are and, or, not



```python
a = True
b = False

# AND
print(a and b)

# OR
print(a or b)

# NOT
print(not a)
```

**Assignment Operator**


*The assignment operator is ( = ). It is used to assign a value to a variable


*   It contain with +=, -=, *=, /=, %=







```python
a = 18
print(a)
a += 5
print(a)
a *= 2
print(a)
a -= 7
print(a)
a /= 9
print(a)
a %= 6
print(a)
```

**Membership Assignment**

*   The membership operators are in and not in. They are used to test whether a value exists within a sequence (like a list, tuple, string, or set).




```python
fruits = ['apple', 'banana', 'cherry']
print('banana' in fruits)
print('orange' in fruits)
print('apple' not in fruits)
print('orange' not in fruits)
print('banana' not in fruits)
```

**Identity Operator**


*   The identity operators are ( is ) and ( is not ). They are used to compare the memory addresses of two objects.




```python
a = 18
b = 7

print(a is b)
print(a is not b)

```

**Bitwise Operator**






*   The Bitwise operators perform operations on binary representations of integers
*   They are AND, OR, XOR, NOT, <<, >>.



```python
a = 5
b = 3
result_1 = a & b
print(result_1)
result_2 = a | b
print(result_2)
result_3 = a ^ b
print(result_3)
result_4 = ~a
print(result_4)
result_5 = a << 1
print(result_5)
result_6 = a >> 1
print(result_6)
```
