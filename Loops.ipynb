**Loop Statement**

  *  The Loop statement allows you to execute a block of code repeatedly
  *  In Loop statement there are for loop and while loop



*  **while loop**
   *   Repeatedly executed a block of code until a condition is met








```python
n = 7
i = 1
while i < n:
  print(i)
  i += 1
```



*   Break terminale to exist the loop
*   continue terminale to skip the loop



```python

count = 0

while count < 10:
    count += 1  # Increment count

    # If count is 5, skip the rest of the loop and continue to the next iteration
    if count == 5:
        continue

    # If count is 8, break out of the loop
    if count == 8:
        break

    print("Count is:", count)
```



*   **for loop**
    *  The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range).



```python
a = "Vinod"
for i in a:
  print(i)
```

```python
# Using for loop with range
for i in range(5):
    print(i + 1)
```
