**Typecasting**


*   It is also called as Type coversion.

*   The procces of changing the data type of a value/object

*  While executing / computaying using operator, there can be mismatching the data type.
*   List item





```python
a = "2"
b = 3
print(a+b) # Error
```

```python
a = "2"
b = 3
print(int(a)+b) #string to integer
```

**Implicit and Expicit Typecasting**


*   Implicit Typecasting

```python
a = 5
print(type(a)) #int
```





*   Explict Typecasting
