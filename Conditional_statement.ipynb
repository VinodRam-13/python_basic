**Conditional statements**


*   It helps to code decisions based on some pre-conditions
*   in conditional statement there are :-

    *   if statement
    *   if-else statement
    *   if-elif-else statement
    *   nested if-else statement



*   **if statement**
    *   The if statement is used for conditional execution of code. It allows  to run a block of code only if a specified condition is true.









```python
x = 10

if x > 5:  # execute if condition is true
    print("x is greater than 5")
```



*   **if-else statement**

    *   In if-else statement it execute the if block when condition is true else block




```python
wheather = "Rainy"
if wheather == "Sunny":
    print("It's a sunny day")
else: #condition is not true, else block will be execute.
    print("It's a rainy day")
```



*   **if-elif-else statement**

    *   the if, elif, and else statements allow to perform conditional execution based on multiple conditions. This structure is useful when need to evaluate several conditions sequentially.







```python
age = 30

if age < 18:
    print("You are a minor.")
elif age >= 18 and age < 65:
    print("You are an adult.")
else:
    print("You are a senior citizen.")
```



*   **nested-if-else statement**

    *   nest if-else statements to create more complex conditional logic. This means placing one if-else statement inside another. It allows for multiple levels of conditions to be checked.




```python
age = 25
citizenship = "INDIA"

if age >= 18:
    if citizenship == "INDIA":
        print("You are eligible to vote in INDIA.")
    else:
        print("You are not eligible to vote in INDIA.")
else:
    print("You must be at least 18 years old to vote.")
```
